# Final Project

The CNN Moneyâ€™s Market Movers website (https://money.cnn.com/data/hotstocks/ ) tracks the most active stocks on a real time basis. Specifically, the most active, the top gainers and top losers are listed at any instance in time. You will first write Python scripts that collect the list of most actives, gainers and losers from the above website. Next, your programs should take the ticker symbols and names of these companies (and categories) and build a csv file (called stocks.csv) with data about each stock from the website: https://finance.yahoo.com/quote/AMD?p=AMD&.tsrc=fin-srch-v1 which gives the quote for ticker symbol AMD as an example. The data to be collected from the Yahoo Finance site should include:
OPEN price
PREV CLOSE price
VOLUME
MARKET CAP

###  Algorithm
- Step 1- import BeautifulSoup library from bs4, requests and csv
- Step 2- create blank csv file where we are going to add data
- Step 3- make a function that will write data to csv file 
- Step 4- Using BeautifulSoup library scape the stock info from two websites
- Step 5- used for loop to write stock info line by line
- Step 6- get the ticker input from user
- Step 7- use csv.reader to read csv file line by line
- Step 8- if ticker in the list it will give open, previous close, volume and market cap data of stock
- Step 9- close file
- Step 10- end

```python
# Import all important libraries from we scrapping and csv writting
from bs4 import BeautifulSoup
import requests
import csv

#First link where we will get stock data
yahoo_url_link = 'https://finance.yahoo.com/quote/stock_data?p=stock_data&.tsrc=fin-srch-v1'
#Open blanck csv file
cnn_stocks = open('stocks.csv','w')
csv_writer = csv.writer(cnn_stocks,lineterminator='\n')

#define a function that will be called later to add stock data line by line
def stock_data(ticker,stock_name):
    
    new_link = yahoo_url_link.replace('stock_data',ticker)
    source_1 = requests.get(new_link).text
    soup_1 = BeautifulSoup(source_1,'lxml')
    
    raw_data = soup_1.find('div',attrs={'id':'quote-summary'})
    final_data = raw_data.find_all('td',class_='Ta(end) Fw(600) Lh(14px)')
    csv_writer.writerow([ticker,stock_name,final_data[1].span.text,final_data[0].span.text,final_data[6].span.text,final_data[8].span.text])

#Using requests and beautifulsoup library scrape data from cnn website
source = requests.get('https://money.cnn.com/data/hotstocks/').text
soup = BeautifulSoup(source,'lxml')

#find_all function to get all table data from HTML. It will give a list
all_stocks = soup.find_all('table')

#first table for most actives
pp = all_stocks[0].find_all('tr')
print('Most Actives:')
csv_writer.writerow(['Most Actives:'])

#first element in list is not important for project, so we will start from second
for e in pp[1:]:
    #get ticker 
    ticker = e.find('td').a.text
    #get stock name
    stock_name = e.find('span').text 
    #write everuthing in csv by calling stock_data function
    stock_data(ticker,stock_name)
    print(ticker, stock_name)
print()
#second table for gainers, everything remains same as above
pp = all_stocks[1].find_all('tr')
print('Gainers:')
csv_writer.writerow(['Gainers:'])
for e in pp[1:]:
    ticker = e.find('td').a.text
    stock_name = e.find('span').text 
    stock_data(ticker,stock_name)
    print(ticker, stock_name)
print()   
#third table for losers, everything remains same as above
pp = all_stocks[2].find_all('tr')
print('Losers:')
csv_writer.writerow(['Losers:'])
for e in pp[1:]:
    ticker = e.find('td').a.text
    stock_name = e.find('span').text 
    stock_data(ticker,stock_name)
    print(ticker, stock_name)   
#close the file, which is always good to do
cnn_stocks.close()

#get user input
user_input = input('Which stock are you interested in:').upper()
print()
#open file in read mode to get stock data
file_open = open('stocks.csv','r')
csv_reader = csv.reader(file_open)
counter =[]

#use for loop to get stock data of user input
for line in csv_reader:
    counter.append(line[0])
    if line[0] == user_input:
        print('The data for',line[0],line[1],'is the following:')
        print(line[0],line[1])
        print('OPEN:', line[2])
        print('PREV CLOSE:', line[3])
        print('VOLUME:', line[4])
        print('MARKET CAP:', line[5])
        break
#if user ticker not in list it will give following line
if user_input not in counter:
    print("Please enter ticker from above list")    
file_open.close()        
```
